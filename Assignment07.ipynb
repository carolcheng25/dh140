```python
from bs4 import BeautifulSoup
```

```python
html_string = response.text
text = BeautifulSoup(html_string, "html.parser")
```

```python
print(text)
```

```python
import nltk
```

```python
from nltk.tokenize import word_tokenize, sent_tokenize
```

```python
sent = sent_tokenize(html_string)
print(sent)
```

```python
words = []
for s in sent:
    for w in word_tokenize(s):
        words.append(w)
```

```python
print(words)
```

```python
from nltk.corpus import stopwords
from string import punctuation
nltk.download('stopwords')
```

```python
print(punctuation)
```

```python
print(stopwords.words('english'))
```

```python
print(list(punctuation))
```

```python
myStopWords = list(punctuation) + stopwords.words('english')
```

```python
print(words)
```

```python
wordsNoStop = []
for i in words:
    if i not in myStopWords:
        wordsNoStop.append(i)
print(words)
print(wordsNoStop)
```

```python
from nltk.stem import PorterStemmer
porter = PorterStemmer()
```

```python
words = [porter.stem(word) for word in words]
```

```python
print(words)
```

```python
from nltk.probability import FreqDist
```

```python
freq = FreqDist(wordsNoStop)
```

```python
freq
```

```python
for i in sorted(freq, key=freq.get, reverse=True)[:20]:
    print(i,freq[i])
```

```python
import pandas as pd
```

```python
from sklearn import datasets
```

```python
df = datasets.load_diabetes(as_frame=True)
```

```python
features_df = df.data
```

```python
target_df = df.target
```

```python
print(df.DESCR)
```

```python
features_df.describe()
```

```python
target_df.describe()
```

```python
import numpy as np
```

```python
for column in features_df.columns:
    print(f"{column}: {np.corrcoef(features_df[column], target_df)[0, 1]}")
```

```python
plt.scatter(
    features_df["bmi"],
    target_df
)
plt.xlabel("bmi")
plt.ylabel("target")
plt.show()
```

The plotted points match up with the correlation coefficient because we can see that the graph is positive and the correlation coefficient is also positi

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
results = model.fit(features_df["bmi"].values.reshape(-1, 1), target_df)
print(results.coef_, results.intercept_)
```

```python
plt.scatter(
    features_df["bmi"],
    target_df
)
plt.plot(
    features_df["bmi"],
    features_df["bmi"] * results.coef_[0] + results.intercept_,
    color="red"
)
plt.xlabel("bmi")
plt.ylabel("target")
plt.show()
```

```python
print(f"r^2 = {model.score(features_df['bmi'].values.reshape(-1, 1), target_df)}")
```

```python
y_pred = features_df["bmi"] * results.coef_[0] + results.intercept_
print(f"MSE = {np.mean((target_df - y_pred)**2)}")
```

```python
plt.scatter(
    features_df["age"],
    target_df
)
plt.xlabel("age")
plt.ylabel("target")
plt.show()
```

```python
print(f"r^2 = {model.score(features_df['age'].values.reshape(-1, 1), target_df)}")
```

```python
y_pred2 = features_df["sex"] * results.coef_[0] + results.intercept_
print(f"MSE = {np.mean((target_df - y_pred2)**2)}")
```
