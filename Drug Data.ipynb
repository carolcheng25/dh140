To see the total alcohol use disorder in the past year within the 12-17 age ranges in 2002, we can use a for loop to find every single time 2002 is mentioned and find the correlating state and alcohol use rates.

```python
for i in report:

    if i['Year'] == 2002:
        print(i['Year'],i['State'],i ['Totals']['Alcohol']['Use Disorder Past Year']['12-17'])
```

We can also use the for loop command to create a function in which we are able to compare the use of cigarettes in the past month in Alabama within the 18-25 age group for any years

```python
def statediff(state1, state2):
    for i in report:
        if i['Year'] == 2010:
            if i['State'] == state1:
                numstate1 = i['Totals']['Marijuana']['New Users']['12-17']
            elif i['State'] == state2:
                numstate2 = i['Totals']['Marijuana']['New Users']['12-17']
    if numstate2 > numstate1:
        print('New Users for Marijuana in 2010 for', state2, 'was greater than for', state1)
    elif numstate1 > numstate2 : 
        print('New Users for Marijuana in 2010 for', state2, 'was less than for', state1)
    else: 
        print('New Users for Marijuana in 2010 for', state2, 'was  equal to that of', state1)
```

```python
statediff('California','Wyoming')
```

```python
statediff('Nevada','Oregon')
```
