```python
# Choose a column that has numeric values and make a line plot of the values
import matplotlib.pyplot as plt
df.loc[df['Year'] == 2010].plot(kind='line', x='State', y='Totals.Alcohol.Use Disorder Past Year.12-17')
```

```python
# Use "loc" to print out the first 10 elements of the plotted column
df.loc[df['Year'] == 2010].iloc[:10,5]
#df.loc[df['Year'] == 2010].loc[408:408+10,'Totals.Alcohol.Use Disorder Past Year.12-17'] gives the same output
```

```python
# Use "loc" to print out the first 10 elements of the plotted column
# as well as the matching 10 elements of a different column that has interesting text
#df.loc[df['Year'] == 2010].iloc[:10].plot(kind='line', x='State', y='Totals.Alcohol.Use Disorder Past Year.12-17')
#df.loc[df['Year'] == 2010].loc[408:408+10,['Totals.Alcohol.Use Disorder Past Year.12-17','State']]
df[['Year','State','Totals.Alcohol.Use Disorder Past Year.12-17']][df['Year']==2010].iloc[:10]
```

```python
# Assign the dataframe values from the previous cell into a new dataframe variable
# and make a bar plot with the text values horizontally and the numeric values as the bar heights
a = df.loc[df['Year'] == 2010].loc[408:408+10,['Totals.Alcohol.Use Disorder Past Year.12-17','State']]
a.plot(kind='bar', x='State')
```

```python
# Re-do the plot from the previous cell as a horizontal bar plot
a.plot(kind='barh', x='State')
```

```python
# Re-do the plot from the previous cell
# and change at least two aesthetic elements (colors, labels, titles, ...)
a.plot(kind='barh', x='State', figsize=(10,10), color='m')
```
