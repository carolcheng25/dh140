```python
response = requests.get('https://www.gutenberg.org/cache/epub/2265/pg2265.txt')
```

```python
from bs4 import BeautifulSoup
```

```python
html_string = response.text
text = BeautifulSoup(html_string, "html.parser")
```

```python
print(text)
```

```python
import nltk
```

```python
from nltk.tokenize import word_tokenize, sent_tokenize
```

```python
sent = sent_tokenize(html_string)
print(sent)
```

```python
words = []
for s in sent:
    for w in word_tokenize(s):
        words.append(w)
```

```python
print(words)
```

```python
from nltk.corpus import stopwords
from string import punctuation
nltk.download('stopwords')
```

```python
print(punctuation)
```

```python
print(stopwords.words('english'))
```

```python
print(list(punctuation))
```

```python
myStopWords = list(punctuation) + stopwords.words('english')
```

```python
print(words)
```

```python
wordsNoStop = []
for i in words:
    if i not in myStopWords:
        wordsNoStop.append(i)
print(words)
print(wordsNoStop)
```

```python
from nltk.probability import FreqDist
```

```python
freq = FreqDist(wordsNoStop)
```

```python
freq
```

```python
for i in sorted(freq, key=freq.get, reverse=True)[:20]:
    print(i,freq[i])
```

```python
listOfMovieWords = []
for i in listOfMovieReviews:
    listOfMovieWords.append([w for w in word_tokenize(i.lower()) if w not in myStopWords])
```

```python
listOfMovieWords
```

```python
from nltk.stem.porter import PorterStemmer
```

```python
p_stemmer = PorterStemmer()
```

```python
listOfStemmedWords = []
for i in listOfMovieWords:
    listOfStemmedWords.append([p_stemmer.stem(w) for w in i])
```

```python
!pip install gensim
```

```python
from gensim import corpora, models
import gensim
```

```python
dictionary = corpora.Dictionary(listOfStemmedWords)
```

```python
print(dictionary.token2id)
```

```python
corpus = [dictionary.doc2bow(text) for text in listOfStemmedWords]
```

```python
ldamodel = gensim.models.ldamodel.LdaModel(corpus, 
                                           num_topics=2, 
                                           id2word = dictionary, 
                                           passes=20)
```

```python
for i in ldamodel.print_topics(num_topics=2, num_words=20):
    print(i)
```
